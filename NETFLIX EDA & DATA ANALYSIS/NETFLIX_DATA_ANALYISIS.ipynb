### **Q11.1  WHAT IS THE MAXIMUM DURATION OF A Movie ON NETLIX**

```python
for i in range(len(Netflix_df)):
  if "min" in Netflix_df["Duration"][i]:
    Netflix_df.loc[i, "Minutes"] = Netflix_df.loc[i, "Duration"].split(" ")[0].split("min")[0]
    Netflix_df.loc[i, "Seasons"] = 0
  elif "Season" in Netflix_df["Duration"][i]:
    Netflix_df.loc[i, "Minutes"] = 0
    Netflix_df.loc[i, "Seasons"] = Netflix_df.loc[i, "Duration"].split(" ")[0].split("Season")[0]
  else:
    Netflix_df.loc[i, "Minutes"] = 0
    Netflix_df.loc[i, "Seasons"] = 0
```

```python
Netflix_df.drop(columns =['Duration'],inplace=True,axis=1)
```

```python
Netflix_df.head(2)
```

```python
Netflix_df['Minutes'] = Netflix_df['Minutes'].astype(int)
```

```python
Netflix_df['Minutes'].max()
```

```python
Netflix_df[Netflix_df['Category']=='Movie'].sort_values(by='Minutes', ascending=False).head(1)
```

### **Q12. WHICH INDIVIDUAL COUNTRY HAS HIGHEST NUMBER OF TV SHOWS**

```python
tv_show = Netflix_df[Netflix_df['Category']=='TV Show']
tv_show
```

```python
tv_show['Country'].value_counts().head(1)
```

```python
df1 = tv_show.groupby('Country').Country.count()
df_country = pd.DataFrame(df1)
df_country.columns=['TV_SHOW_COUNT']
df_country.index.name = 'Country'
```

```python
df_country=df_country.sort_values(by='TV_SHOW_COUNT',ascending=False)
# df_country
```

```python
top_5 = df_country.head(5)
top_5
```

```python
plt.figure(figsize=(14,7))
sns.barplot(data=top_5 , x=top_5.index , y='TV_SHOW_COUNT', palette='Paired')
plt.show()
```
